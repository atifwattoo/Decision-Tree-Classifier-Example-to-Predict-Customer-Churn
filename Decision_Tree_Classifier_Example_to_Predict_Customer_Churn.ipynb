```python
#import necessary liberaries
import pandas as pd
import matplotlib.pyplot as plt
import warnings
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn import tree

warnings.filterwarnings("ignore")
```

```python
#Creating a synthetic dataset
#This dataset simulate customer dataset for Telecom company

data = {
    'CustomerID': range(1, 101),  # Unique ID for each customer
    'Age': [20, 25, 30, 35, 40, 45, 50, 55, 60, 65]*10,  # Age of customers
    'MonthlyCharge': [50, 60, 70, 80, 90, 100, 110, 120, 130, 140]*10,  # Monthly bill amount
    'CustomerServiceCalls': [1, 2, 3, 4, 0, 1, 2, 3, 4, 0]*10,  # Number of customer service calls
    'Churn': ['No', 'No', 'Yes', 'No', 'Yes', 'No', 'Yes', 'Yes', 'No', 'Yes']*10  # Churn status
}

# Create a DataFrame from the synthetic data
# for key, values in data.items():
#   print(f"{key}: {values}")

df = pd.DataFrame(data)  # Create a DataFrame from the synthetic data
# print(df)
print(df.head()) # The head() function in Pandas is used to display the first few rows of a DataFrame,
# print(df.tail()) #The tail() function in Pandas is used to display the last few rows of a DataFrame.

```

```python
# Splitting the dataset into features and target variable
# Features include age, monthly charges and customer service calls
# Target variable include Churn (Yes or Not)

X = df[['Age', 'MonthlyCharge', 'CustomerServiceCalls']] #Features
y = df['Churn'] # Target Variable

# Splittig the dataset into training and testing set
# 70% data is used for training and 30% is used for testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Training the Decision Tree Modek
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)

# Making Prediction on the Test Set
y_pred = clf.predict(X_test)

# Evaluating the Model Using accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
```

```python
# Visualizing the decision tree
# This visualizing helps to understand how model makes decisions
plt.figure(figsize=(12, 8))
tree.plot_tree(clf, filled=True, feature_names=['Age', 'MonthlyCharge', 'CustomerServiceCalls'] , class_names= ['no churn', 'churn'])
plt.show()
```
